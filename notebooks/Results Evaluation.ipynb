# Plot Training Curves

```python
df = pd.read_csv('../logs/seg_net_skip_concatenate_adam_lr_01_block_4_training_2018-11-16 18:30:48.049061.log', delimiter=',')

ax = plt.gca()
ax.set_xmargin(0)
ax.set_ymargin(0)
plt.plot(df['epoch'], df['loss'], label="Train Loss")
plt.plot(df['epoch'], df['val_loss'], label="Validation Loss")
plt.legend()
plt.title('Train and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

# Load Model

```python
json_file = open('../weights/seg_net_skip_concatenate_adam_lr_01_block_4_model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)

# load weights into new model
loaded_model.load_weights("../weights/seg_net_skip_concatenate_adam_lr_01_block_4_weights.60-0.03.hdf5")
print("[INFO] Loaded model from disk.")
```

```python
loaded_model.summary()
```

# Test the model

```python
image_name = '2007_004712'
image = cv2.imread('../dataset/test/input/' + image_name + '.jpg') * 1./255
segmets = cv2.imread('../dataset/test/segments/' + image_name + '.png')
image = cv2.resize(image, (320,320), cv2.INTER_NEAREST)

batch = np.expand_dims(image, axis=0)
label = loaded_model.predict(batch, 1)[0]

plt.imshow(image[...,::-1])
```

```python
label_reshape = label.reshape((320,320,21))
t = np.argmax(label_reshape, axis=2)
```

## Save w/ color palette

```python
img = Image.fromarray(t.astype(np.uint8)).convert('P')
img.putpalette(PALETTE)
img.save(OUTPUT_DIR + image_name + '.png')
```
